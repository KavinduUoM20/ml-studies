**b)  Do you think standardizing the variables before clustering would be a good idea? Justify your answer.**

Yes. because Standardizing,

*1. Ensures Equal Weighting -* Prevents variables with larger ranges from dominating distance calculations, leading to more balanced clustering.

*2. Improves Algorithm Performance -* Enhances the effectiveness of distance-based algorithms (e.g., K-Means, Hierarchical clustering, DBSCAN) by achieving more meaningful clusters.

*3. Facilitates Comparison -* Allows straightforward comparison of each variable's contribution to clusters, especially for features with different units or ranges.

*4. Prevents Bias -* Avoids undue influence of larger scale variables on cluster centroids, resulting in unbiased clustering outcomes.

### **Import Dependancies**

```python
import pandas as pd
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from scipy.cluster.hierarchy import linkage, dendrogram, fcluster
from sklearn.metrics import adjusted_rand_score, silhouette_score
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv('./data/planet.csv')
```

### **2) Standardizing**

```python
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
```

### **3) Hierarchical Clustering**

```python
# Hierarchical clustering
linkage_matrix = linkage(df_scaled, method='complete', metric='euclidean')
```

```python
# Visualize
plt.figure(figsize=(10, 6))
dendrogram(linkage_matrix)
plt.title('Dendrogram of Exoplanets')
plt.xlabel('Exoplanets')
plt.ylabel('Euclidean Distance')
plt.show()
```

```python
# Extract three clusters
clusters = fcluster(linkage_matrix, 3, criterion='maxclust')
```

```python
# Assign cluster labels
df['HCresults'] = clusters
```

```python
# Calculate cluster-specific means
cluster_means = df.groupby('HCresults').mean()
cluster_means
```

```python
# Pairwise scatterplots
sns.pairplot(df, hue='HCresults', palette='Set1')
plt.suptitle('Pairwise Scatterplots')
plt.show()
```

### **4) K-Means Clustering**

```python
# K-means clustering with K=3
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(df_scaled)
```

```python
# Assign cluster labels to the original DataFrame
df['KMCresults'] = kmeans.labels_
```

```python
# cluster-specific means 
cluster_means = df.groupby('KMCresults').mean()
cluster_means 
```

```python
# Pairwise scatterplots 
sns.pairplot(df, hue='KMCresults', palette='Set1')
plt.suptitle('Pairwise Scatterplots')
plt.show()
```

### **5) Comparing Results**

```python
df.head()
```

```python
hierarchical_clusters = df['HCresults']
kmeans_clusters = df['KMCresults']
```

#### Adjusted Rand Index (ARI)

```python
# Adjusted Rand Index (ARI)
ari = adjusted_rand_score(hierarchical_clusters, kmeans_clusters)
print(f"Adjusted Rand Index (ARI): {ari}")
```

**Interpretation**

ARI of 0.173 indicates a low to moderate level of agreement between the clustering results from hierarchical clustering and K-means clustering.

It suggests that there is some similarity between the clusters identified by the two methods, but there are also significant differences or inconsistencies.

#### Silhouette Score

```python
# Silhouette Score for K-means clustering
silhouette_score_kmeans = silhouette_score(df_scaled, kmeans_clusters)
print(f"Silhouette Score (K-means): {silhouette_score_kmeans}")
```

```python
# Silhouette Score for Hierarchical clustering
silhouette_score_hierarchical = silhouette_score(df_scaled, hierarchical_clusters)
print(f"Silhouette Score (Hierarchical): {silhouette_score_hierarchical}")
```

**Interpretation**
  
The Silhouette Score for K-means clustering (0.337) suggests that the clusters identified by K-means have a moderate level of separation and coherence.

The Silhouette Score for hierarchical clustering (0.252) indicates a slightly lower level of cluster separation compared to K-means.


#### Conclusion

K-means clustering appears to have slightly better-defined clusters, as indicated by its higher Silhouette Score.
