## One Hot Encoding

```python
import pandas as pd
```

```python
data = pd.read_csv('./data/ohe-homeprices.csv')
```

```python
data.head()
```

```python
data.shape
```

```python
#Get dummy variables
dummies = pd.get_dummies(data.town, dtype=int)
dummies
```

```python
merged = pd.concat([data,dummies],axis = 'columns')
merged
```

```python
final = merged.drop(['town','west windsor'],axis='columns')
final
```

We drop one dummy variable column to avoid being vulnerable to the Dummy Variable Trap (Which is due to occurance of Multicollinearity)

```python
from sklearn import linear_model
```

```python
reg = linear_model.LinearRegression()
```

```python
#X/Y Split
X = final.drop('price',axis='columns')
X
```

```python
y = final.price
y
```

```python
reg.fit(X,y)
```

```python
reg.predict([[2800,0,1]])
```

```python
reg.predict([[2800,0,0]])
```

```python
reg.predict([[2800,1,0]])
```

```python
reg.score(X,y)
```
