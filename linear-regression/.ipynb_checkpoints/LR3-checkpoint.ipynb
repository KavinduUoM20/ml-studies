## Multiple Linear Regression Basics

#### 1. Import Dependancies

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
```

#### 2. Load Dataset

```python
data = pd.read_csv('./data/mul-homeprices.csv')
```

```python
data.head()
```

```python
data.shape
```

#### 3. Data Cleaning

```python
data.columns[data.isnull().any()]
```

```python
bedrooms_median = int(data['bedrooms'].median())
bedrooms_median
```

```python
data.bedrooms = data.bedrooms.fillna(bedrooms_median)
```

```python
data.head()
```

#### 4. Fit Regression Model

```python
reg = linear_model.LinearRegression()
reg.fit(data[['area','bedrooms','age']],data.price)
```

```python
reg.coef_
```

```python
reg.intercept_
```

#### 5. Predictions

```python
reg.predict([[5000,2,10]])
```
